## a. In a jupyter notebook markdown cell, describe the difference between linear and rank correlation.

__LINEAR CORRELATION:__ measures the strength of association between two variables (x and y and the direction of the relationship AS THEY RUN IN A STRAIGHT LINE. +1 and -1 both mean there is a relationship, while 0 means there is none. 
   * has interval or ratio level - as the independent variable increases or decreases the function increases or decreases
   * straight line

__RANK CORRELATOIN__ - measures the strength between two variables that run in a line that is a curved pattern, also known as monotonic relationship 
  *  as the independent variable increases the function neither increases nor decreases in tandem
  *  wavey line or curved line

```python
## b. Create a scatter plot showing the relationship between calculated_host_listings_count
##      and availability_365 in the Belize Airbnb data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv ('belize_airbnb_listings.csv')
y = data['calculated_host_listings_count']
x = data['availability_365']
colors = np.random.rand(2796)

print(np.corrcoef(x,y))

plt.scatter(x,y, c = colors)
plt.title('Correlation between calculated host_listings and availability_365')
plt.ylabel('calculated_host_listings')
plt.xlabel('availability_365')
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)), color='black')
plt.show()
```

 With a 0.17 there is barely a positive association between the amount of times that a host listed their B and B, and the availability of the property. In short, the association, though positive, is very small.
