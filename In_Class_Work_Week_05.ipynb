1. Make a function that can take any non-negative integer as an argument and return it with its digits in descending order. Essentially, rearrange the digits to create the highest possible number. If no number is passed in, make the default value inputted = 1111111

Input = 43905  Output = 95430

Input = 584366 Output = 866543

```python
def order(n):
    if type(n) != int:
        print('111111')
    elif n <= 0:
            print("Nope, enter a positive integer that is more than 0")
    else:
        descending = "".join(sorted(str(n), reverse=True)) #changes int to string and uses the string to sort
        print(descending)

order(43905)
order(584366)
order("hi")
order(-1)
```
